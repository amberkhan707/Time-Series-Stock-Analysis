# ML Start

Test Train data split

```python
startdate = datetime(1949,1,1)
enddate = datetime(1955,12,1)
teststart = datetime(1956,1,1)
testend = datetime(1960,12,1)
```

```python
train = df[startdate : enddate]
test = df[teststart : testend]
```

ADF on Train to check Stationary 

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def sol(series):
    result = adfuller(series)
    print("ADF-Statistics", result[0])
    print("P-Value", result[1])
    if result[1] < 0.05:
        print("Data is stationary")
    else :
        print("Data is not Stationary")
```

```python
train['2d'] = train['1d'] - train['1d'].shift(1)
```

```python
sol(train['month'].dropna())
```

We understood our D value is 2

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
acf = plot_acf(train['2d'].dropna())
acf
```

Q=1

```python
pacf = plot_pacf(train['2d'].dropna())
pacf
```

p = 11

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
Arimamodel = ARIMA(train['Thousands of Passengers'], order=(11, 2, 3),enforce_stationarity=False, enforce_invertibility=False)
Arimamodel_fit = Arimamodel.fit()
Arimamodel_fit.summary()
```

```python
predict = Arimamodel_fit.predict(test.index[0],test.index[-1])
```

```python
Arimamodel_fit.resid.plot(kind = 'kde')
plt.show()
```

```python
test['prediction'] = predict
```

```python
ax = test[['Thousands of Passengers','prediction']].plot(figsize = (12,6))
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

```python
test['error'] = abs(test['Thousands of Passengers'] - test['prediction'])
```

```python
test['accuracy'] = (1 - (test['error'] / test['Thousands of Passengers'])) * 100
```

```python
test[['accuracy']].mean()
```

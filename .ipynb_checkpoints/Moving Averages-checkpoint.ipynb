```python
import yfinance as yf 
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = yf.download('TSLA')
```

```python
df.head()
```

```python
 df.columns = df.columns.droplevel(1)
```

# Simple Moving Average 

```python
df['200SMA'] = df.rolling(200,1).mean()['Open']
```

```python
df[['200SMA','50SMA','Open']].plot(xlim = ['2020-01-01','2025-01-01'],figsize=(12,5))
plt.show()
```

```python
df.head()
```

# Cumulative Moving Average

```python
 df['Open'].expanding().mean().plot()
plt.show()
```

# Exponential Weightage Moving Average 
1. Isme apne hissab se smoothing factor dete h

```python
df['50EMA'] = df['Open'].ewm(50).mean()
df['200EMA'] = df['Open'].ewm(200).mean()
```

```python
df[['Open','50EMA','200EMA']].plot(figsize = (12,6),xlim = ['2020-01-01','2024-01-01'])
plt.show()
```
